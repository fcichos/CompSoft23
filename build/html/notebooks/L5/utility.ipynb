```python
import numpy as np
import matplotlib.pyplot as plt 

%matplotlib inline
%config InlineBackend.figure_format = 'retina' 

plt.rcParams.update({'font.size': 12,
                     'axes.titlesize': 18,
                     'axes.labelsize': 16,
                     'axes.labelpad': 14,
                     'lines.linewidth': 1,
                     'lines.markersize': 10,
                     'xtick.labelsize' : 16,
                     'ytick.labelsize' : 16,
                     'xtick.top' : True,
                     'xtick.direction' : 'in',
                     'ytick.right' : True,
                     'ytick.direction' : 'in',}) 
```

```python
x=np.linspace(0,2*np.pi,100)
xl=np.linspace(0,2*np.pi,10)
y=np.sin(x)
yl=np.sin(xl)
```

```python
plt.figure(figsize=(8,6))
plt.bar(xl[:-1],yl[:-1],fill=False,align='edge',width=0.7,lw=2)
plt.plot(x,y,'k--',lw=1)
plt.xlabel('x')
plt.ylabel('y')
plt.tight_layout()


plt.savefig('img/box.png')
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.plot(xl,yl,'k-',lw=2)
plt.axhline(y=0,color='k')
plt.vlines(xl,ymin=0,ymax=yl,lw=2)
plt.plot(x,y,'k--',lw=1)
plt.xlabel('x')
plt.ylabel('y')
plt.tight_layout()


plt.savefig('img/trapez.png')
plt.show()
```
